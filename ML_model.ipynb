```python
import pandas as pd
df = pd.read_csv('model_data.xls')
df
```

```python
import warnings

import matplotlib
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

y = df['steam_review_percent']
X_raw = df.drop('steam_review_percent', axis=1)
normalizer = MinMaxScaler()

X_real_norm_np = normalizer.fit_transform(X_raw)
X = pd.DataFrame(data=X_real_norm_np)
X.head()
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
lr = LinearRegression().fit(X_train, y_train)
print(f"Train: {mean_squared_error(y_train, lr.predict(X_train))}")
print(f"Test: {mean_squared_error(y_test, lr.predict(X_test))}")
```

```python
ridge = Ridge(1).fit(X_train, y_train)
print("\nRidge")
print(f"Train: {mean_squared_error(y_train, ridge.predict(X_train))}")
print(f"Test: {mean_squared_error(y_test, ridge.predict(X_test))}")
```
